USING BETA, PE AND EPS AS A MEASURE OF VOLATILITY TO ASSIGN STOCKS TO INVESTOR'S GROUP ACCORDING TO THEIR RISK PROFILE

Beta: It's a measure of a stock's volatility in relation to the overall market. A beta of 1 means the stock tends to move with the market, higher than 1 means it's more volatile, and less than 1 means it's less volatile.

PE (Price-to-Earnings ratio): It's the ratio of a company's share price to its earnings per share. It indicates how much investors are willing to pay per dollar of earnings. A high PE might indicate that investors expect higher earnings growth in the future.

EPS (Earnings Per Share): It's a company's profit allocated to each outstanding share of common stock. It's a portion of a company's profit and is a crucial measure of a company's profitability.

```python
import matplotlib.pyplot as plt

# Assigning categories to the stocks
conservative_stocks['Category'] = 'Conservative'
moderate_stocks['Category'] = 'Moderate'
aggressive_stocks['Category'] = 'Aggressive'

# Concatenating the categorized dataframes
categorized_stocks = pd.concat([conservative_stocks, moderate_stocks, aggressive_stocks])

# Plotting
plt.figure(figsize=(8, 6))

colors = {'Conservative': 'green', 'Moderate': 'blue', 'Aggressive': 'red'}

for category, color in colors.items():
    category_stocks = categorized_stocks[categorized_stocks['Category'] == category]
    plt.scatter(category_stocks['Beta'], category_stocks['PE'], c=color, label=category, alpha=0.7)

plt.xlabel('Beta')
plt.ylabel('PE')
plt.title('Categorized Stocks based on Beta and PE')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Prepare the data
features = categorized_stocks[['Beta', 'PE', 'EPS']]
target = categorized_stocks['Category']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Initialize the model
model = LogisticRegression()

# Fit the model
model.fit(X_train, y_train)

# Make predictions
predictions = model.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```

An accuracy score of 0.5555 implies that your model is correctly predicting the categories for approximately 55.55% of the test data. This metric reflects the ratio of correctly predicted categories to the total test set. In other words, about 55.55% of the time, your model accurately classified stocks into their appropriate categories based on the provided features (Beta, PE, and EPS).

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Function to fetch stock data
def get_stock_evolution(stock_symbols):
    data = yf.download(stock_symbols, period="1y")
    return data

# List of stock symbols
portfolio_symbols = ['ADBE', 'CAT', 'DPZ', 'META', 'FDX', 'IT', 'GS', 'HD', 'JPM', 'MCD']

# Fetch stock data
stock_evolution = get_stock_evolution(portfolio_symbols)

initial_investment = 500
monthly_investment = 100

investments = [initial_investment]  # Track the investments over time
portfolio_value = [initial_investment]  # Track the portfolio value over time

# Calculate investment and portfolio values
for i in range(1, len(stock_evolution)):
    # Increment investment each month
    investment = investments[-1] + monthly_investment
    investments.append(investment)

    # Calculate the portfolio value
    value = 0
    for stock_symbol in portfolio_symbols:
        start_price = stock_evolution['Adj Close'][stock_symbol][0]
        trading_days = len(stock_evolution)  # Total trading days in the dataset
        current_price_index = min(i * 5 - 1, trading_days - 1)  # Considering 5 trading days in a week
        current_price = stock_evolution['Adj Close'][stock_symbol][current_price_index]
        value += (monthly_investment / start_price) * current_price

    portfolio_value.append(value)

# Plotting the investment and portfolio value over time
plt.figure(figsize=(10, 6))
plt.plot(portfolio_value, label='Portfolio Value', marker='o')
plt.plot(investments, label='Investment', marker='o')
plt.xlabel('Days')
plt.ylabel('Value')
plt.title('Investment vs Portfolio Value')
plt.legend()
plt.grid(True)
plt.show()


```
